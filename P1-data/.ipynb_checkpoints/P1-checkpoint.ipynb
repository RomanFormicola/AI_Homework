```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

age = []
Hours_per_week = []
for row in data:
    if len(row) > 0:
        age.append(row[0])
    if len(row) > 12:
        Hours_per_week.append(row[12])

age.sort()
plt.hist(age, bins=10)
plt.xticks(np.arange(0, 90, step=10))
plt.xlabel("age")
plt.ylabel("frequence(#)")
plt.show()

Hours_per_week.sort()
plt.hist(Hours_per_week, bins=5)
plt.xticks(np.arange(0, 90, step=10))
plt.xlabel("Hours Per Week")
plt.ylabel("frequence(#)")
plt.show()

                          
```

**ii)**

```python
ageLessThen50k = []
ageGreaterThen50k = []

hoursLessThen50k = []
hoursGreaterThen50k = []

for row in data:
    if len(row) > 14:
        if(row[14].strip() == "<=50K"):
            ageLessThen50k.append(row[0])
            hoursLessThen50k.append(row[12])
        elif(row[14].strip() == ">50K"):
            ageGreaterThen50k.append(row[0])
            hoursGreaterThen50k.append(row[12])
        
ageLessThen50k.sort()
ageGreaterThen50k.sort()
hoursLessThen50k.sort()
hoursGreaterThen50k.sort()

plt.xticks(np.arange(0, 90, step=10))
plt.xlabel("age")
plt.ylabel("frequence(#)")
plt.title("Age Compared by Income >50k or <=50k using 1990 Census Data")
plt.hist(ageLessThen50k, bins=10, alpha=0.5, label='<=50k')
plt.hist(ageGreaterThen50k, bins=10, alpha=0.5, label='>50k')
plt.legend(loc='upper right')
plt.show()

plt.xticks(np.arange(0, 90, step=10))
plt.xlabel("Hours Per Week")
plt.ylabel("frequence(#)")
plt.title("Hours Worked Per Week Compared by Income >50k or <=50k using 1990 Census Data")
plt.hist(hoursLessThen50k, bins=10, alpha=0.5, label='<=50k')
plt.hist(hoursGreaterThen50k, bins=10, alpha=0.5, label='>50k')
plt.legend(loc='upper right')
plt.show()

```

**iii)**

```python
for i in range(len(ageLessThen50k)):
    ageLessThen50k[i] = int(ageLessThen50k[i])
for i in range(len(ageGreaterThen50k)):
    ageGreaterThen50k[i] = int(ageGreaterThen50k[i])

ageSideBySide = [ageLessThen50k, ageGreaterThen50k]
    
fig = plt.figure(figsize =(10, 7)) 
ax = fig.add_axes([0, 0, 1, 1]) 
bp = ax.boxplot(ageSideBySide)

ax.set_xticklabels(['<=50k', '>50k']) 
plt.title("Age Compared by Income >50k or <=50k using 1990 Census Data")
plt.ylabel("age")
plt.show()


for i in range(len(hoursLessThen50k)):
    hoursLessThen50k[i] = int(hoursLessThen50k[i])
for i in range(len(hoursGreaterThen50k)):
    hoursGreaterThen50k[i] = int(hoursGreaterThen50k[i])
    
hoursSideBySide = [hoursLessThen50k, hoursGreaterThen50k]
    
fig = plt.figure(figsize =(10, 7)) 
ax = fig.add_axes([0, 0, 1, 1]) 
bp = ax.boxplot(hoursSideBySide)

ax.set_xticklabels(['<=50k', '>50k']) 
plt.title("Hours Worked Per Week Compared by Income >50k or <=50k using 1990 Census Data")
plt.ylabel("Hours Per Week")
plt.show()
```

**iv)**
The age histograms have revealed that the majority of americans make less then 50k a year and the majority of americans regardless of income work between 35 and 55 hours per week. The median hours per week for both th >50k and <=50k groups is very close to 40 hours per week. The median age for people making less the 50k a year is around 35 while the median age for people making greater then 50k a year is around 45 years old. A higher percentage of people who make >50k a year work more then 45 hours per week as compared to the percentage of people who make <=50k a year who work more then 45 hours per week.

```python
from collections import Counter

educationDict = {}
maritalDict = {}


for row in data:
    if len(row) > 14:
        if row[3].strip() not in educationDict.keys():
            educationDict[row[3].strip()] = 1
        else:
            educationDict[row[3].strip()] += 1
            
        if row[5].strip() not in maritalDict.keys():
            maritalDict[row[5].strip()] = 1
        else:
            maritalDict[row[5].strip()] += 1

plt.figure(figsize=(25,12))
plt.bar(educationDict.keys(), educationDict.values())
plt.title("Highest Educational Achievement Among U.S. Citizens 1990 Census ")
plt.ylabel("# of People")
plt.xlabel('Educational Achievement')
plt.show()

plt.figure(figsize=(25,12))
plt.bar(maritalDict.keys(), maritalDict.values())
plt.title("Marital Status Among U.S. Citizens 1990 Census ")
plt.ylabel("# of People")
plt.xlabel('Marital Status Achievement')
plt.show()

        
```

**ii)**

```python
eduDictLess50K = {}
eduDictGreat50K = {}

maritalDictLess50K = {}
maritalDictGreat50K = {}

for row in data:
    if len(row) > 14:
        if(row[14].strip() == "<=50K"):
            if row[3].strip() not in eduDictLess50K.keys():
                eduDictLess50K[row[3].strip()] = 1
            else:
                eduDictLess50K[row[3].strip()] += 1
            if row[5].strip() not in maritalDictLess50K.keys():
                maritalDictLess50K[row[5].strip()] = 1
            else:
                maritalDictLess50K[row[5].strip()] += 1 
                
        elif(row[14].strip() == ">50K"):
            if row[3].strip() not in eduDictGreat50K.keys():
                eduDictGreat50K[row[3].strip()] = 1
            else:
                eduDictGreat50K[row[3].strip()] += 1
            if row[5].strip() not in maritalDictGreat50K.keys():
                maritalDictGreat50K[row[5].strip()] = 1
            else:
                maritalDictGreat50K[row[5].strip()] += 1
                
plt.figure(figsize=(25,12))
plt.bar(eduDictLess50K.keys(), eduDictLess50K.values(), alpha=0.5, label='<=50K')
plt.bar(eduDictGreat50K.keys(), eduDictGreat50K.values(), alpha=0.5, label='>50K')
plt.title("Highest Educational Achievement Among U.S. Citizens Comparing people with income >50k vs <=50k per year")
plt.ylabel("# of People")
plt.xlabel('Educational Achievement')
plt.legend(loc='upper right')
plt.show()

plt.figure(figsize=(25,12))
plt.bar(maritalDictLess50K.keys(), maritalDictLess50K.values(), alpha=0.5, label='<=50K')
plt.bar(maritalDictGreat50K.keys(), maritalDictGreat50K.values(), alpha=0.5, label='>50K')

plt.title("Marital Status Among U.S. Citizens Comparing people with income >50k vs <=50k per year")
plt.ylabel("# of People")
plt.xlabel('Marital Status')
plt.legend(loc='upper right')
plt.show()

            
```

**iii**) From the bar charts we can conclude that U.S. adults in 1990 who make greater then 50k a year are much more likely to be married than people who make less and vice versa. U.S. adults in 1990 who are widowed are also more likely to make less then 50k a year. U.S. adults in 1990 who earn greater then 50k a year are also more like then people who make less then 50k a year to have earned doctorates or bachelors degrees.

```python
data = []
with open("nfl-20-running-stats.csv") as fData_file:
    fData_reader = csv.reader(fData_file, delimiter=',')
    for row in fData_reader:
        if row[4].strip().upper() == 'RB' or row[4].strip().upper() == 'FB':
            if int(row[5].strip()) >= 6:
                row[4] = row[4].strip().upper()
                row[5] = int(row[5].strip())
                data.append(row)
    
```

```python
print("Size of Pool:", len(data))
```

**b)**

```python
import statistics
TD = []
fmb = []
yds = []
firstDown = []
YG = []
YA = []
Lng = []
for player in data:
    if(len(player) == 15):
        TD.append(int(player[9]))
        fmb.append(int(player[14]))
        yds.append(int(player[8]))
        firstDown.append(int(player[10]))
        YG.append(float(player[13]))
        Lng.append(int(player[11]))
        YA.append(float(player[12]))

TD.sort()
fmb.sort()
yds.sort()
firstDown.sort()
YG.sort()
Lng.sort()

print("Mean TD:", statistics.mean(TD))
print("Median TD:", statistics.median(TD))
print("Mode TD:", statistics.mode(TD))
print()
print("Mean Fmb:", statistics.mean(fmb))
print("Median Fmb:", statistics.median(fmb))
print("Mode Fmb:", statistics.mode(fmb))
print()
```
